```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('energies.data', sep=' ', header=None)
bands = 6
```

```python
# First column is the k-values
kvals = df[0]
# All other columns are the eigenvalues
eigenvalues = df.drop(0, axis=1)
energies = eigenvalues.values
```

```python
# Plot the eigenvalues

for n in range(bands):
    plt.scatter(kvals, energies[:,n], c="black", s=2)
plt.show()
```
